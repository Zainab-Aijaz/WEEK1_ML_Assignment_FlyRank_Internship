```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Pages:", len(df))
print("Clients:", df["client_id"].nunique())

declining_share = (df["trend_direction"] == "down").mean()
print(f"Pages trending down (last 30d vs prev 30d): {declining_share:.1%}")

zero_click_share = (df["clicks_last_30d"] == 0).mean()
print(f"Pages with zero clicks in the last 30 days: {zero_click_share:.1%}")

decay_risk = (df["avg_position"] <= 10) & (df["trend_direction"] == "down")
print(f"Page-one pages (avg position <=10) that are declining: {decay_risk.mean():.1%}")
```

*Across 30,000 pages from 32 clients: 54.2% are trending down in clicks over the last 30 days vs. the prior 30. 61.2% got zero clicks in the last 30 days — a huge pool nobody would think to manually check one by one. Most strikingly, 24.4% of page-one pages (avg position ≤10) are actively declining — these are the highest-value pages to lose quietly, since they're already winning real estate and could slip off page one before anyone notices. That combination — high volume, mostly-invisible signal, real cost if missed — is exactly the "worth 7 weeks" case for this lane.*
